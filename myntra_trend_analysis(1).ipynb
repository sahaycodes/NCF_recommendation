```python
from bs4 import BeautifulSoup
from urllib.parse import urlparse, urljoin
import ast
import requests
from PIL import Image
from io import BytesIO
import os
from IPython.display import display, Image as IPImage

# Read the csv file into a DataFrame
d1_df = pd.read_csv(r'cleaned_myntra_shreya.csv')

# Function to extract the image URL from the 'Product Image' column
#def extract_image_url(product_images):
    #try:
        #image_list = ast.literal_eval(product_images)
        #return image_list[0]['src'] if image_list and 'src' in image_list[0] else ''
    #except (ValueError, SyntaxError):
        #return ''

# Apply the function to the 'Product Image' column and create a new column 'Image URL'
#d1_df['Image URL'] = d1_df['URL'].apply(extract_image_url)

# Ensure the 'images' directory exists
os.makedirs('images', exist_ok=True)

def extract_images_from_url(url, product_id):
    try:
        response = requests.get(url)
        if response.status_code == 200:
            soup = BeautifulSoup(response.content, 'html.parser')
            img_urls = []

            # Extract image URLs from <img> tags
            for img_tag in soup.find_all('img'):
                img_url = img_tag.get('src')
                if img_url:
                    # Convert relative URLs to absolute URLs
                    img_url = urljoin(url, img_url)
                    img_urls.append(img_url)

            # Download images
            for img_url in img_urls:
                try:
                    img_name = urlparse(img_url).path.split('/')[-1]
                    img_path = f'images/{product_id}_{img_name}'
                    img_data = requests.get(img_url).content
                    with open(img_path, 'wb') as f:
                        f.write(img_data)
                    print(f"Image saved: {img_path}")
                except Exception as e:
                    print(f"Error downloading image {img_url}: {e}")

        else:
            print(f"Error accessing URL {url}: Status code {response.status_code}")

    except Exception as e:
        print(f"Error accessing URL {url}: {e}")

# Apply the function to each row in the DataFrame
for index, row in d1_df.iterrows():
    url = row['URL']
    product_id = row['Product_id']
    if url.strip():
        print(f"Processing {url} for product_id {product_id}")
        extract_images_from_url(url, product_id)
    else:
        print(f"Empty URL for product_id {product_id}")


# Function to download and save images locally
def download_image(url, product_id):
    if url.strip():
        try:
            response = requests.get(url)
            if response.status_code == 200:
                img = Image.open(BytesIO(response.content))
                img_path = f'images/{product_id}.jpg'
                img.save(img_path)
                return img_path
            else:
                print(f"error downloading {url}: Status code{response.status_code}")
                return None
        except Exception as e:
            print(f"Error downloading {url}: {e}")
            return None
    else:
        print(f"Empty URL for product_id {product_id}")
        return None

# Download images and save the file paths
d1_df['Image Path'] = d1_df.apply(lambda row: download_image(row['Image URL'], row['Product_id']), axis=1)

# Function to display images in a DataFrame
def display_image(img_path):
    try:
        return IPImage(filename=img_path)
    except:
        return None

# Apply the display function to the 'Image Path' column
d1_df['Display Image'] = d1_df['Image Path'].apply(display_image)

# Display the DataFrame with images in a Jupyter notebook
d1_df[['Product_id', 'Image URL', 'Image Path', 'Display Image']].head()

# If running in a non-interactive environment, you can use display to show images manually
for img_path in d1_df['Image Path'].dropna().head(10):  # Display the first 10 images
    display(IPImage(filename=img_path))
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Embedding, Flatten, Dense, concatenate

# Load your DataFrame
user_item_data = pd.read_csv(r'cleaned_myntra_shreya.csv')

# Neural Collaborative Filtering
user_encoder = LabelEncoder()
products_encoder = LabelEncoder()
category_encoder = LabelEncoder()
gender_encoder = LabelEncoder()
brand_encoder = LabelEncoder()

def create_ncf_model(user_item_data, embedding_size=50):
    user_item_data['user'] = user_encoder.fit_transform(user_item_data['Product_id'])
    user_item_data['products'] = products_encoder.fit_transform(user_item_data['Product_id'])
    user_item_data['category'] = category_encoder.fit_transform(user_item_data['Category'])
    user_item_data['gender'] = gender_encoder.fit_transform(user_item_data['category_by_Gender'])
    user_item_data['brand'] = brand_encoder.fit_transform(user_item_data['BrandName'])

    num_users = user_item_data['user'].nunique()
    num_products = user_item_data['products'].nunique()
    num_categories = user_item_data['category'].nunique()
    num_genders = user_item_data['gender'].nunique()
    num_brands = user_item_data['brand'].nunique()

    # Model architecture
    user_input = Input(shape=(1,), name='user')
    products_input = Input(shape=(1,), name='products')
    category_input = Input(shape=(1,), name='category')
    gender_input = Input(shape=(1,), name='gender')
    brand_input = Input(shape=(1,), name='brand')
    rating_input = Input(shape=(1,), name='Ratings')

    user_embeddings = Embedding(input_dim=num_users, output_dim=embedding_size, name='user_embedding')(user_input)
    products_embeddings = Embedding(input_dim=num_products, output_dim=embedding_size, name='products_embedding')(products_input)
    category_embeddings = Embedding(input_dim=num_categories, output_dim=embedding_size, name='category_embedding')(category_input)
    gender_embeddings = Embedding(input_dim=num_genders, output_dim=embedding_size, name='gender_embedding')(gender_input)
    brand_embeddings = Embedding(input_dim=num_brands, output_dim=embedding_size, name='brand_embedding')(brand_input)

    user_vecs = Flatten()(user_embeddings)
    products_vecs = Flatten()(products_embeddings)
    category_vecs = Flatten()(category_embeddings)
    gender_vecs = Flatten()(gender_embeddings)
    brand_vecs = Flatten()(brand_embeddings)

    input_vecs = concatenate([user_vecs, products_vecs, category_vecs, gender_vecs, brand_vecs, rating_input])
    x = Dense(128, activation='relu')(input_vecs)
    x = Dense(64, activation='relu')(x)
    y = Dense(1)(x)

    model = Model(inputs=[user_input, products_input, category_input, gender_input, brand_input, rating_input], outputs=y)
    model.compile(optimizer='adam', loss='mean_squared_error')

    return model

ncf_model = create_ncf_model(user_item_data)
ncf_model.summary()

# Train the model
ncf_model.fit(
    [user_item_data['user'], user_item_data['products'], user_item_data['category'], user_item_data['gender'], user_item_data['brand'], user_item_data['Ratings']],
    user_item_data['Ratings'],
    epochs=10,
    verbose=1
)
```
